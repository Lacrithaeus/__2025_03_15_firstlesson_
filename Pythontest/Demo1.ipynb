```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2π (covering one full sine wave)
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = sin(x)', color='b', linewidth=2)

# Highlight key points (0, π/2, π, 3π/2, 2π) for better understanding
key_x = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
key_y = np.sin(key_x)
labels = ['0', 'π/2', 'π', '3π/2', '2π']

# Plot the key points
plt.scatter(key_x, key_y, color='red', zorder=3)
for i in range(len(key_x)):
    plt.text(key_x[i], key_y[i] + 0.1, labels[i], fontsize=12, ha='center')

# Add dashed x-axis for reference
plt.axhline(0, color='gray', linestyle='dashed')

# Add labels and title
plt.xlabel('x (radians)')
plt.ylabel('y')
plt.title('Understanding the Sine Wave')

# Add grid and legend
plt.grid(True, linestyle='dotted')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values from 0 to 2π with 1000 points
x = np.linspace(0, 2 * np.pi, 1000)
y = np.cos(x)

# Key points for better understanding
key_x = np.array([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
key_y = np.cos(key_x)
key_labels = ["0", "π/2", "π", "3π/2", "2π"]

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, linestyle="--", color="blue", label="cos(x)")

# Mark key points
plt.scatter(key_x, key_y, color="red", zorder=3)
for i in range(len(key_x)):
    plt.text(key_x[i], key_y[i] + 0.1, key_labels[i], fontsize=12, ha="center")

# Add labels and title
plt.xlabel("x (radians)")
plt.ylabel("cos(x)")
plt.title("Understanding the Cosine Wave")
plt.axhline(0, color="black", linewidth=1)  # X-axis
plt.axvline(0, color="black", linewidth=1)  # Y-axis

# Set x-ticks at key points
plt.xticks(key_x, key_labels)

# Show the grid and legend
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Display the plot
plt.show()
```
